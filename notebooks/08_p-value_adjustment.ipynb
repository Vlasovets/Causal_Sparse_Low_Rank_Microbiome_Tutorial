```python
import pandas as pd
import pickle
import plotly.express as px
import statsmodels.stats.multitest as smt
import rpy2.robjects as ro

from rpy2.robjects.packages import importr
from rpy2.robjects import pandas2ri, FloatVector
from rpy2.robjects.conversion import localconverter
# Convert pandas.DataFrames to R dataframes automatically.
pandas2ri.activate()

from utils.helper import r_to_pandas
```

```python
utils = importr('utils')
devtools = importr('devtools')
fdrtool = importr('fdrtool')

n_iter = 10000
level = "species"
dataset = "not_only"
```

### LinDA

#### Assymptotic

```python
# with open('../data/linda_assymptotic_pvalues.pkl', 'rb') as f:
#     linda_assymp = pickle.load(f)
    
with open('data/linda_assymptotic_pvalues_ige_{0}.pkl'.format(dataset), 'rb') as f:
    linda_assymp = pickle.load(f)
    
pd.DataFrame(linda_assymp['species'].sort_values())
```

#### Randomized

```python
# p_value = pd.read_csv('data/adj_linda_{0}_{1}_age_sex_bmi.csv'.format(level, n_iter)) # ASV_TABLE_1469 
p_value = pd.read_csv('data/adj_linda_{0}_{1}_ige_{2}.csv'.format(level, n_iter, dataset))
p_value = pd.DataFrame(linda_assymp['species']).sort_values(by=['pvalue'])
p_value = p_value.rename(columns={'pvalue': 'p-value'})
p_value
```

```python
#should be uniform
fig_pvalue = px.histogram(p_value['p-value'], x="p-value", nbins = 30, color_discrete_sequence=['#641E62'])
fig_pvalue.write_image("plots/svg/linda_pvalue_agg.svg")
fig_pvalue.show()
```

```python
bh = smt.multipletests(p_value['p-value'], method='fdr_bh')
```

```python
p_value['BH'] = bh[1]
```

```python
fdr = fdrtool.fdrtool(p_value["p-value"], statistic="pvalue")
```

```python
local_fdr = r_to_pandas(fdr.rx2("qval"))
```

```python
p_value['local_FDR'] = local_fdr
```

```python
p_value.head(20)
```
